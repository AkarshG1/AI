```python
import gym
env = gym.make('CartPole-v0')

#> Discrete(2)
print(env.observation_space)

for i in range(100):
    env.reset()
    obs=r=done=info=None
    while done!=True:
        obs,r,done,info = env.step(1)
    print(f'obs={obs},r={r},done={done},i={info}')
env.close()
```
